```python
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
import urllib.request

#January 2025

resource_id = 'EPD_202501'
total_rows = 5000
batch_size = 1000
records = []

# Loop to get more data
for offset in range(0, total_rows, batch_size):
    url = f'https://opendata.nhsbsa.net/api/3/action/datastore_search?resource_id={resource_id}&limit={batch_size}&offset={offset}'
    with urllib.request.urlopen(url) as response:
        content = response.read()
    data = json.loads(content.decode("utf-8"))
    batch_records = data["result"]["records"]
    records.extend(batch_records)
```

```python
data = json.loads(content.decode("utf-8"))
records = data["result"]["records"]
df = pd.DataFrame(records)
```

```python
df.info
```

```python
df['NIC'] = pd.to_numeric(df['NIC'], errors='coerce')
df['ACTUAL_COST'] = pd.to_numeric(df['ACTUAL_COST'], errors='coerce')
df['ITEMS'] = pd.to_numeric(df['ITEMS'], errors='coerce')
```

```python
df_clean = df[df['REGIONAL_OFFICE_NAME'] != 'UNIDENTIFIED']
```

```python
df_clean['NIC_per_item'] = df_clean['NIC'] / df_clean['ITEMS']
```

```python
region_sum = df_clean.groupby('REGIONAL_OFFICE_NAME').agg({
    'NIC': 'sum',
    'ITEMS': 'sum'
}).reset_index()
```

```python
region_sum['NIC_per_item'] = region_sum['NIC'] / region_sum['ITEMS']
```

```python
print(region_sum)
```

```python
#Regional Cost for Prescribtion
plt.figure(figsize=(10,6))
plt.bar(region_sum['REGIONAL_OFFICE_NAME'], region_sum['NIC_per_item'])
plt.xticks(rotation=45)
plt.ylabel('Average NIC per Item')
plt.title('Regional Average NIC per Item')
plt.tight_layout()
plt.show()
```

```python
#Expensive Drugs
top_expensive = df.groupby('CHEMICAL_SUBSTANCE_BNF_DESCR')['NIC'].sum().sort_values(ascending=False).head(10)
top_expensive.plot(kind='barh', figsize=(10,6), title='Top 10 Drugs by Total NIC')
plt.xlabel('Total NIC')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
drug_name = "Semaglutide"  # Checking the drug prescription
drug_df = df[df['CHEMICAL_SUBSTANCE_BNF_DESCR'].str.contains(drug_name, case=False)]

regional_use = drug_df.groupby('REGIONAL_OFFICE_NAME')['ITEMS'].sum().sort_values(ascending=False)
regional_use.plot(kind='bar', figsize=(10,6), title=f'{drug_name} Prescriptions by Region')
plt.ylabel('Items')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Hospital-Level Prescribing
practice_summary = df[df['PRACTICE_NAME'] != 'UNIDENTIFIED'].groupby('PRACTICE_NAME').agg({
    'ITEMS': 'sum',
    'NIC': 'sum',
    'ACTUAL_COST': 'sum'
}).sort_values(by='ITEMS', ascending=False)

practice_summary.head(10)
```

```python
#integrated Care Boards
icb_sum = df[df['ICB_NAME'] != 'UNIDENTIFIED'].groupby('ICB_NAME').agg({
    'NIC': 'sum',
    'ACTUAL_COST': 'sum',
    'ITEMS': 'sum'
}).reset_index()

icb_sum['NIC_per_item'] = icb_sum['NIC'] / icb_sum['ITEMS']
icb_sum.sort_values(by='NIC_per_item', ascending=False).head(10)
```

```python
plt.figure(figsize=(12,6))
plt.barh(icb_sum.sort_values(by='NIC_per_item', ascending=False)['ICB_NAME'].head(10),
         icb_sum.sort_values(by='NIC_per_item', ascending=False)['NIC_per_item'])
plt.xlabel("NIC per Item")
plt.title("Top 10 ICBs by NIC per Item")
plt.tight_layout()
plt.show()
```
